## Descriptive statistics

Descriptive statistics is a fundamental aspect of data science and statistics. They are used to summarize and describe the basic features of a dataset, giving us a quick insight into the sample and measures of the data. Unlike inferential statistics, descriptive statistics don't aim to make conclusions beyond the data we have, nor do they try to infer about a population from a sample. Instead, they are all about describing what's in front of us.

> Any single summary statistic, while useful, doesn't tell the full story of a dataset. Each statistic compresses the information of a variable into a single representative number. However, collectively, these statistics provide us with valuable information about the data's distribution,variability and central tendencies,

## Key Elements of Descriptive Statistics

### Measures of Central Tendency
> They give us an idea about the typical or central value in the dataset.
- **Mean (Average)**: The sum of all values divided by the number of values.
  - Balance point: The mean can be visualized as the balance point where the number line would perfectly balance if each data point had the same weight placed at its respective location.
![balance_point_image](https://www.gastonsanchez.com/matrix4sl/images/mean/mean-balancing.png)

- **Median**: The middle value when the data is ordered.
- **Mode**: The most frequently occurring value(s).
  - Mode is rarely an informative statistic about the central tendency of the data. It's most useful in describing the **typical** observation of a categorical variable.


### Measures of Spread
> These measures tell us how spread out or tightly clustered the data points are.
- **Range**: The difference between the highest and lowest values.
- **Variance**: A measure of how much the values in a dataset are spread out.
- **Standard Deviation**: The average distance of each data point from the mean.
- **Interquartile Range**: The difference between the 25th and 75th percentiles.

### Measures of Shape
> These are about the distribution of data,
- **Skewness**: How asymmetric the data distribution is.
- **Kurtosis**: How 'peaked' the data distribution is compared to a normal distribution.

### Other Descriptive Measures
- **Count**: The number of data points.
- **Minimum and Maximum Values**: The smallest and largest values.
- **Quartiles**: Values that divide the data into four equal parts.




## Example

Imagine you are a teacher and have the final grades for a class of 20 students. The grades are:  56, 62, 67, 70, 71, 72, 73, 75, 76, 78, 78, 80, 80, 82, 85, 88, 90, 92, 94, 97.


````python
import numpy as np

grades = np.array([56, 62, 67, 70, 71, 72, 73, 75, 76, 78, 78, 80, 80, 82, 85, 88, 90, 92, 94, 97])
````

````python
# Calculate the mean (average) of the grades
# Formula: Mean = Sum of all values / Number of values
mean = np.mean(grades)
````

````python
# Calculate the median of the grades
# Formula: Median is the middle value when the data is ordered
median = np.median(grades)
````

````python
from scipy import stats

# Calculate the mode of the grades
# Formula: Mode is the value that appears most frequently
mode = stats.mode(grades)
````

````python
# Calculate the range of the grades
# Formula: Range = Maximum value - Minimum value
range_ = np.ptp(grades)
````

````python
# Calculate the variance of the grades
# Formula: Variance = Sum of squared differences from the mean / Number of values
variance = np.var(grades)
````

````python
# Calculate the standard deviation of the grades
# Formula: Standard Deviation = Square root of the variance
std_dev = np.std(grades)
````

````python
# Print the results
print(f"Mean (Average): {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode.mode[0]:.2f}")
print(f"Range: {range_:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
````

````python
skewness = stats.skew(grades)
kurtosis = stats.kurtosis(grades)

print(f"skewness: {skewness:.2f}")
print(f"kurtosis: {kurtosis:.2f}")
````

````python
import matplotlib.pyplot as plt
import seaborn as sns


# Setting up the subplots
plt.figure(figsize=(15, 10))

# Histogram
plt.subplot(2, 2, 1)
sns.histplot(grades, kde=True, color='skyblue')
plt.axvline(mean, color='r', linestyle='-', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='g', linestyle='--', label=f'Median: {median:.2f}')
plt.title('Histogram of Grades')
plt.legend()

# Boxplot
plt.subplot(2, 2, 2)
sns.boxplot(x=grades, color='lightgreen')
plt.title('Box Plot of Grades')

# Summary statistics text
plt.subplot(2, 2, 3)
plt.axis('off')
plt.text(0, 0.9, f'Range: {range_}\nVariance: {variance:.2f}\nStandard Deviation: {std_dev:.2f}',
         fontsize=12, ha='left')
plt.text(0, 0.8, f'Skewness: {skewness:.2f}\nKurtosis: {kurtosis:.2f}', fontsize=12, ha='left')
plt.title('Descriptive Statistics')

# Show the plots
plt.tight_layout()
plt.show()
````

# Statistical Moments

Statistical moments are key measures that describe various characteristics of a probability distribution. They are used for both populations and samples.

## Population Moments
Population moments are calculated using the entire data from a population and provide a theoretical understanding of the distribution's characteristics.

- **First Moment (Mean):** The average value of the dataset.
- **Second Moment (Variance):** Measures the spread of the data around the mean.
- **Third Moment (Skewness):** Indicates the asymmetry of the distribution around the mean.
- **Fourth Moment (Kurtosis):** Describes the "tailedness" or the peak of the distribution.

**N-th Moment about the Origin (Population):**
$$
\mu'_n = E(X^n) = \int_{-\infty}^{\infty} x^n f(x) dx
$$
- $\mu'_n$: n-th raw moment.
- $E(X^n)$: Expected value of $X^n$.
- $f(x)$: Probability density function.

**N-th Central Moment (Population):**
$$
\mu_n = E[(X - \mu)^n] = \int_{-\infty}^{\infty} (x - \mu)^n f(x) dx
$$
- $\mu_n$: n-th central moment.
- $\mu$: Population mean.

## Sample Moments
Sample moments are calculated using data from a sample, a subset of the population. They provide practical estimates of the population moments.

- **First Moment (Sample Mean):** The average value of the sample.
- **Second Moment (Sample Variance):** Measures the spread of the sample data around the sample mean.
- **Third Moment (Sample Skewness):** Indicates the asymmetry of the sample distribution around the sample mean.
- **Fourth Moment (Sample Kurtosis):** Describes the "tailedness" or the peak of the sample distribution.

**N-th Moment about the Origin (Sample):**
$$
\overline{x}_n = \frac{1}{N} \sum_{i=1}^{N} x_i^n
$$
- $\overline{x}_n$: n-th raw moment for the sample.
- $x_i$: Sample data points.
- $N$: Sample size.

**N-th Central Moment (Sample):**
$$
m_n = \frac{1}{N} \sum_{i=1}^{N} (x_i - \overline{x})^n
$$
- $m_n$: n-th central moment for the sample.
- $\overline{x}$: Sample mean.


**Mean**:

* **High Mean**: Indicates that the average value of the data points is large.
* **Low Mean**: Indicates that the average value of the data points is small.


**Variance**:

* **High Variance**: Suggests a wide spread of data points around the mean; the data points vary significantly from each other.
* **Low Variance**: Suggests a narrow spread; the data points are very similar to each other and close to the mean.

**Skewness**:
* **Skewness Close to Zero**: Implies that the data distribution is fairly symmetrical around the mean.
* **High Positive Skewness**: Indicates a long tail on the right side, with data clustered to the left (mean is less than the median).
* **High Negative Skewness**: Indicates a long tail on the left side, with data clustered to the right (mean is greater than the median).

**Kurtosis**:
* **Kurtosis Close to Zero (when compared to a normal distribution, kurtosis = 3)**: Indicates a distribution with tail thickness similar to a normal distribution (mesokurtic).
* **High Kurtosis (>3)**: Indicates heavy tails and a sharp peak (leptokurtic); more prone to outliers and extreme values.
* **Low Kurtosis (<3)**: Indicates light tails and a flatter peak (platykurtic); less prone to outliers and extreme values.

**General Note**:
Values close to zero for skewness and kurtosis (with kurtosis adjusted for the normal distribution baseline of 3) suggest that the distribution is close to normal in terms of symmetry and tail thickness.


The `scipy.stats.kurtosis` function calculates the excess kurtosis by default. Excess kurtosis is obtained by subtracting 3 from the kurtosis of the distribution, which aligns the measure with the kurtosis of a normal distribution:

- **Positive Excess Kurtosis (>0):** The distribution has heavier tails than a normal distribution (leptokurtic).
- **Negative Excess Kurtosis (<0):** The distribution has lighter tails than a normal distribution (platykurtic).
- **Zero Excess Kurtosis (0):** The distribution has tails similar to those of a normal distribution (mesokurtic).

This standardization by subtracting 3 allows for a direct comparison of tail thickness relative to the normal distribution:

```python
# To calculate the excess kurtosis
excess_kurtosis = scipy.stats.kurtosis(data)

# To calculate the non-excess (absolute) kurtosis
absolute_kurtosis = scipy.stats.kurtosis(data, fisher=False)


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Normal distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Positively skewed distribution
positive_skew_data = np.random.exponential(scale=2, size=1000)

# Negatively skewed distribution
negative_skew_data = -np.random.exponential(scale=2, size=1000)

# Calculate skewness and kurtosis
print("Normal Distribution Skewness:", skew(normal_data))
print("Normal Distribution Kurtosis:", kurtosis(normal_data))

print("Positive Skew Distribution Skewness:", skew(positive_skew_data))
print("Positive Skew Distribution Kurtosis:", kurtosis(positive_skew_data))

print("Negative Skew Distribution Skewness:", skew(negative_skew_data))
print("Negative Skew Distribution Kurtosis:", kurtosis(negative_skew_data))

# Plot histograms
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, color='blue', edgecolor='black')
plt.title('Normal Distribution')

plt.subplot(1, 3, 2)
plt.hist(positive_skew_data, bins=30, color='green', edgecolor='black')
plt.title('Positive Skew Distribution')

plt.subplot(1, 3, 3)
plt.hist(negative_skew_data, bins=30, color='red', edgecolor='black')
plt.title('Negative Skew Distribution')

plt.tight_layout()
plt.show()
````
